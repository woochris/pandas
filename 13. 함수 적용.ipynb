## 데이터에 함수 적용 (apply)

```python
# 키 뒤에 cm을 붙이는 함수
def add_cm(height) :
    return str(height) + 'cm'

df['키'] = df['키'].apply(add_cm) # 키 데이터에 대해서 add_cm 함수를 호출한 결과 데이터 반영
df
```

```python
def cap(lang) : 
    if pd.notnull(lang) : # NaN이 아닌지 if문
        return lang.capitalize() #첫 글자는 대문자, 나머지 글자는 소문자로
    return lang

df['SW특기'] = df['SW특기'].apply(cap)
df
```
