## 데이터 제외하기 dropna

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df.dropna(inplace=True) # 전체 데이터 중에서 NaN을 포함하는 데이터 삭제
```

```python
df
```

```python
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

- axis : index or columns  # index는 row를 지움, columns은 column을 지움
- how : any or all  # any는 NaN이 하나라도 있으면 삭제, all은 전체데이터가 NaN이여야 삭제

```python
df.dropna(axis='index', how='any') # NaN이 하나라도 있는 row를 삭제
```

```python
df.dropna(axis='columns', how='any') # how='any'는 디폴트값
```

```python
df['학교'] = np.nan
df
```

```python
df.dropna(axis='columns', how='all') # column 데이터 전체가 NaN일 경우에만 column 삭제
```
